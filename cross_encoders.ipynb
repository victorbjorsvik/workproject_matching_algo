```python
jobs = sorted(os.listdir("job_descriptions"))[:6]
jobs
```

```python
for job in jobs:
    print("="*50)
    description_file_path = os.path.join('job_descriptions', job)
    with open(description_file_path, 'r') as file:
        job_description = file.read()
    df_jobs = pd.DataFrame([job_description], columns=["raw"])
    df_jobs = main.job_info_extraction(df_jobs)
    print(job)
    print("="*50)
    print("job skills:")
    print(df_jobs["Skills"][0])
    print("-"*50)
    bi_encoder = main.calc_similarity(df_resumes, df_jobs, parallel=True)
    cross_encoder = main.calc_cross(df_resumes, df_jobs, parallel=True)
    print("old approach:")
    print("-"*50)
    print(bi_encoder.sort_values("rank"))
    print("-"*50)
    print("new approach:")
    print("-"*50)
    print(cross_encoder.sort_values("softmaxed", ascending=False))
    
```
